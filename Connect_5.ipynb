### Connect 5



*   O jogo deve ser jogado entre 2 pessoas em turnos, cada um com peças de uma cor.
*   O objetivo consiste em formar uma linha, coluna ou diagonal construídas por CINCO peças adjacentes de mesma cor. O jogador que conseguir tal formação primeiro é considerado vitorioso.
*   Caso não seja possível fazer mais jogadas antes que se obtenha um jogador vencedor, o jogo é considerado empatado.
*   Em seu turno, o jogador deve colocar uma peça de modo ADJACENTE a outra peça qualquer já posicionada no tabuleiro. Isto é, ela deve conectar-se verticalmente, horizontalmente ou pela diagonal com outra peça qualquer no tabuleiro. Outras jogadas são consideradas inválidas. 
*  O turno do jogador termina ao fazer uma jogada válida.





## Tabuleiro

> O tabuleiro "vazio" padrão é um tabuleiro 10 por 10, com os quatro quadrados centrais já preenchidos com peças (duas de cada cor). Essas peças devem ser dispostas de modo que formem um “X”, isto é, peças de mesma cor conectam-se pela diagonal. Veja a figura abaixo.

grid.png



> Podemos também aumentar ou diminuir o tamanho do tabuleiro de modo que siga o padrão  de um tabuleiro N por N para N par, assim, as peças centrais continuariam centralizadas no tabuleiro. No entanto, um tabuleiro muito pequeno como o 4x4 por exemplo, invalidaria o jogo, pois não seria possível conectar 5 peças de mesma cor.

```
def get_valid_moves(tab) :
  valids = []
  for i in range(len(tab)) :
    n=0
    if tab[i] == 0 :

        # Direita      
        try:
          if tab[i+1]!= 0:
            n = n+1
        except IndexError:
          pass

        # Esquerda
        try:
          if tab[i-1]!= 0:
            n = n+1
        except IndexError:
          pass

        # Abaixo
        try:
          if tab[i+10]!= 0:
            n = n+1
        except IndexError:
          pass

        # Acima
        try:
          if tab[i-10]!= 0:
            n = n+1
        except IndexError:
          pass

        # Diagonal Inferior Direita
        try:
          if tab[i+11]!= 0:
            n = n+1
        except IndexError:
          pass

        # Diagonal Inferior Esquerda
        try:
          if tab[i+9]!= 0:
            n = n+1
        except IndexError:
          pass

        # Diagonal Superior Esquerda
        try:
          if tab[i-11]!= 0:
            n = n+1
        except IndexError:
          pass

        # Diagonal Superior Direita
        try:
          if tab[i-9]!= 0:
            n = n+1
        except IndexError:
          pass        
         
        if n != 0:
          valids.append(i)
  return valids
```

## Jogador BOT Bobinho

> Esse jogador irá fazer jogadas aleatoreamente. Ele receberá um tabuleiro e um parâmetro turn para saber de quem é a vez de jogar. Recebido isso, retornará uma jogada ao acaso.

```
## BOT bobinho que faz jogadas ao acaso.

import random

def player(tab, turn) :
    valid_moves = get_valid_moves(tab)
    if valid_moves :
        return (turn, random.choice(valid_moves) )
```
